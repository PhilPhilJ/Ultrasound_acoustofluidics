### Sedimantation velocities

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
xs = np.array([0, 140, 60, 90, 190, 25, 45, 90, 15, 170])
vs = np.array([1.370, 0.8644, 0.9704, 0.9080, 1.185, 1.487, 1.667, 1.049, 1.419, 0.5671])
mean_v = np.mean(vs)
sd_v = np.std(vs)
```

```python
R = 200
def func(r, u_max):
    return u_max * (1-(r/R)**2)

popt, pcov = curve_fit(func, xs, vs)

x = np.linspace(np.min(xs)-10, 200, 10000)
y = func(x, popt)
```

```python
plt.plot(x,y, color="tab:green", linestyle= "dashdot")
plt.scatter(xs, vs, color="black")
plt.xlabel("Distance from center of capillary tube")
plt.ylabel(r"v [$\mathrm{\mu m \,s^{-1}}$]")
plt.title("Sedimentation velocities")
plt.hlines(y=mean_v, xmin=np.min(xs)-10, xmax=np.max(xs)+10)
plt.hlines(y=mean_v + 2*sd_v, xmin=np.min(xs)-10, xmax=np.max(xs)+10, color= "tab:red", linestyle = "dashed")
plt.hlines(y=mean_v - 2*sd_v, xmin=np.min(xs)-10, xmax=np.max(xs)+10, color= "tab:red", linestyle = "dashed")
plt.hlines(y=mean_v + 2*sd_v, xmin=np.min(xs)-10, xmax=np.max(xs)+10, color= "tab:red", linestyle = "dashed")
plt.hlines(y=mean_v - 2*sd_v + 0.02, xmin=100, xmax=110, color= "black")
plt.hlines(y=mean_v - 0.02, xmin=100, xmax=110, color= "black")
plt.hlines(y=mean_v + 2*sd_v - 0.02, xmin=100, xmax=110, color= "black")
plt.hlines(y=mean_v + 0.02, xmin=100, xmax=110, color= "black")
plt.vlines(x=105, ymin=mean_v + 0.02, ymax=mean_v + 2*sd_v -0.02, color="black")
plt.vlines(x=105, ymin=mean_v - 2*sd_v +0.02, ymax=mean_v - 0.02, color="black")

plt.text(-8, 1.05, s="Mean velocity =" + str(round(mean_v,2)) + " " + r"$\mathrm{\mu m \,s^{-1}}$", color="tab:blue")
plt.text(108, 1.5, s=r"2$\mathrm{\sigma}$", color="black")
plt.text(108, 0.8, s=r"2$\mathrm{\sigma}$", color="black")
plt.xlim(np.min(xs)-10, np.max(xs)+10)
plt.ylim(0, 2*mean_v)
plt.legend(["Fitted flow profile"])

plt.savefig(r'C:\Users\Phili\OneDrive - Danmarks Tekniske Universitet\Bachelorprojekt\Images for Overleaf\sedimentation velocities.png', dpi=350,bbox_inches = "tight")
```
