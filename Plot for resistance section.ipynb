```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Q = np.linspace(0.01, 5, 1000)
R_hyd = 11.5
p = Q * R_hyd
```

```python
R_2 = 81
Q_2 = p / R_2
```

```python
R_1 =  13.4
Q_1 = p / R_1
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.plot(Q, p)
ax1.set_title("Pressure difference")
ax1.set(xlabel=r'$ Q\quad [\mathrm{\mu Ls^{-1}}$] ', ylabel=r'$\Delta p\quad [\mathrm{Pa}$]')

ax2.plot(p, Q_1, color="r")
ax2.set_title("Flow rate over $\mathrm{R_1}$")
ax2.set(xlabel=r'$\Delta p\quad [\mathrm{Pa}$]', ylabel=r'$Q \quad [\mathrm{\mu Ls^{-1}}$]')

ax3.plot(p, Q_2, color="r")
ax3.set_title("Flow rate over $\mathrm{R_2}$")
ax3.set(xlabel=r'$\Delta p\quad [\mathrm{Pa}$]', ylabel="")
ax3.sharey(ax2)


plt.subplots_adjust(wspace=0.4)
plt.savefig(r'C:\Users\Phili\OneDrive - Danmarks Tekniske Universitet\Bachelorprojekt\Images for Overleaf\pressure and flow.png')
```
