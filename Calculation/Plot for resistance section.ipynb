```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Q = np.linspace(1, 100, 1000)
R_hyd = 11.5
p = Q * R_hyd
```

```python
R_2 = 81
Q_2 = p / R_2
```

```python
R_1 =  13.4
Q_1 = p / R_1
```

```python
fig, ax = plt.subplots(1, 2, figsize=(5, 5))
ax[0].plot(Q, p)
ax[0].set(xlabel=r'$ Q\quad [\mathrm{\mu Ls^{-1}}$] ', ylabel=r'$\Delta p\quad [\mathrm{Pa}$]')
ax[0].vlines(x=5, ymin=0, ymax=57, color="tab:red",linestyles='dashed')
ax[0].legend([r"$\Delta p $ system"],loc="upper left")
ax[0].set_xlim(1,8)
ax[0].set_ylim(0,100)

ax[1].plot(p, Q_1, color="black")
ax[1].set(xlabel=r'$\Delta p\quad [\mathrm{Pa}$]', ylabel=r'$Q \quad [\mathrm{\mu Ls^{-1}}$]')
ax[1].plot(p, Q_2, color="tab:gray")
ax[1].yaxis.set_label_position("right")
ax[1].hlines(y=5, xmin=p[0], xmax=405, color="tab:red",linestyles='dashed')
ax[1].legend([r"$Q_{\mathrm{By pass}}$", r"$Q_{\mathrm{Capillary}}$"],loc="upper right")
ax[1].vlines(x=68, ymin=0, ymax=5, color="tab:green",linestyles='dashdot')
ax[1].vlines(x=405, ymin=0, ymax=5, color="tab:green",linestyles='dashdot')
ax[1].set_xlim(10,450)
ax[1].set_ylim(0,9)

plt.savefig(r'C:\Users\Phili\OneDrive - Danmarks Tekniske Universitet\Bachelorprojekt\Images for Overleaf\pressure and flow.png', dpi=300,bbox_inches = "tight")
```
